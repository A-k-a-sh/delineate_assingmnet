```python
import os
import json
import random
from pathlib import Path
from typing import Dict, List, Tuple

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image
from tqdm import tqdm
import cv2

# Set style
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 8)
plt.rcParams['font.size'] = 10

# Paths
IMAGES_DIR = "/kaggle/input/med-real-dataset/dataset/images"
LABELS_DIR = "/kaggle/input/med-real-dataset/dataset/labels"

print(f"Images directory: {IMAGES_DIR}")
print(f"Labels directory: {LABELS_DIR}")
```

## 1. Dataset Overview

```python
# Get all files
image_files = sorted([f for f in os.listdir(IMAGES_DIR) if f.endswith('.png')])
label_files = sorted([f for f in os.listdir(LABELS_DIR) if f.endswith('.json')])

print(f"Total images: {len(image_files)}")
print(f"Total labels: {len(label_files)}")
print(f"\nSample files:")
for i in range(min(3, len(image_files))):
    print(f"  {image_files[i]}")

# Verify pairing
image_ids = set([os.path.splitext(f)[0] for f in image_files])
label_ids = set([os.path.splitext(f)[0] for f in label_files])
print(f"\nImages without labels: {len(image_ids - label_ids)}")
print(f"Labels without images: {len(label_ids - image_ids)}")
```

## 2. Load and Parse All Labels

```python
def load_all_labels(labels_dir: str) -> Dict:
    """Load all label files and extract statistics."""
    data = {
        'image_ids': [],
        'lines_per_image': [],
        'points_per_image': [],
        'points_per_line': [],
        'error_bar_up': [],
        'error_bar_down': [],
        'error_bar_total': [],
        'point_coords': [],
        'line_names': [],
        'zero_bars': 0  # Count points with zero-length error bars
    }
    
    for label_file in tqdm(label_files, desc='Loading labels'):
        img_id = os.path.splitext(label_file)[0]
        label_path = os.path.join(labels_dir, label_file)
        
        try:
            with open(label_path, 'r') as f:
                label = json.load(f)
        except:
            continue
        
        data['image_ids'].append(img_id)
        data['lines_per_image'].append(len(label))
        
        total_points = 0
        for line in label:
            line_name = line['label'].get('lineName', 'unknown')
            data['line_names'].append(line_name)
            
            # Actual data points
            actual_points = [p for p in line['points'] if p.get('label', '') == '']
            data['points_per_line'].append(len(actual_points))
            total_points += len(actual_points)
            
            for point in actual_points:
                data['point_coords'].append((point['x'], point['y']))
                
                up = point.get('topBarPixelDistance', 0)
                down = point.get('bottomBarPixelDistance', 0)
                
                data['error_bar_up'].append(up)
                data['error_bar_down'].append(down)
                data['error_bar_total'].append(up + down)
                
                if up == 0 and down == 0:
                    data['zero_bars'] += 1
        
        data['points_per_image'].append(total_points)
    
    return data

data = load_all_labels(LABELS_DIR)

print(f"\nDataset Statistics:")
print(f"  Total images: {len(data['image_ids'])}")
print(f"  Total data points: {len(data['error_bar_up'])}")
print(f"  Total lines: {len(data['points_per_line'])}")
print(f"  Points with zero-length bars: {data['zero_bars']} ({data['zero_bars']/len(data['error_bar_up'])*100:.1f}%)")
```

## 3. Image Characteristics

```python
# Analyze all images
resolutions = []
aspect_ratios = []
file_sizes = []

for img_file in tqdm(image_files, desc='Analyzing images'):
    img_path = os.path.join(IMAGES_DIR, img_file)
    img = Image.open(img_path)
    w, h = img.size
    
    resolutions.append((w, h))
    aspect_ratios.append(w / h)
    file_sizes.append(os.path.getsize(img_path) / 1024)  # KB

widths = [r[0] for r in resolutions]
heights = [r[1] for r in resolutions]

print(f"\nImage Characteristics:")
print(f"  Width:  {np.min(widths)} - {np.max(widths)} px (mean: {np.mean(widths):.0f})")
print(f"  Height: {np.min(heights)} - {np.max(heights)} px (mean: {np.mean(heights):.0f})")
print(f"  Aspect ratio: {np.min(aspect_ratios):.2f} - {np.max(aspect_ratios):.2f} (mean: {np.mean(aspect_ratios):.2f})")
print(f"  File size: {np.min(file_sizes):.0f} - {np.max(file_sizes):.0f} KB (mean: {np.mean(file_sizes):.0f})")
print(f"  Unique resolutions: {len(set(resolutions))}")

# Visualize resolution distribution
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].scatter(widths, heights, alpha=0.6, s=50, c=aspect_ratios, cmap='viridis')
axes[0].set_xlabel('Width (pixels)')
axes[0].set_ylabel('Height (pixels)')
axes[0].set_title('Image Resolutions')
axes[0].grid(alpha=0.3)
cbar = plt.colorbar(axes[0].collections[0], ax=axes[0])
cbar.set_label('Aspect Ratio')

axes[1].hist(aspect_ratios, bins=30, edgecolor='black', alpha=0.7, color='steelblue')
axes[1].axvline(np.median(aspect_ratios), color='red', linestyle='--', linewidth=2, label=f'Median: {np.median(aspect_ratios):.2f}')
axes[1].set_xlabel('Aspect Ratio (width/height)')
axes[1].set_ylabel('Frequency')
axes[1].set_title('Aspect Ratio Distribution')
axes[1].legend()
axes[1].grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.savefig('real_resolutions.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 4. Distribution Analysis

```python
fig, axes = plt.subplots(2, 3, figsize=(18, 10))

# 1. Lines per image
axes[0, 0].hist(data['lines_per_image'], bins=np.arange(0.5, max(data['lines_per_image'])+1.5, 1), 
                edgecolor='black', alpha=0.7, color='steelblue')
axes[0, 0].set_xlabel('Number of Lines per Image')
axes[0, 0].set_ylabel('Frequency')
axes[0, 0].set_title(f'Lines per Image\n(mean: {np.mean(data["lines_per_image"]):.2f})')
axes[0, 0].grid(axis='y', alpha=0.3)

# 2. Points per image
axes[0, 1].hist(data['points_per_image'], bins=30, 
                edgecolor='black', alpha=0.7, color='coral')
axes[0, 1].set_xlabel('Number of Data Points per Image')
axes[0, 1].set_ylabel('Frequency')
axes[0, 1].set_title(f'Points per Image\n(mean: {np.mean(data["points_per_image"]):.1f})')
axes[0, 1].grid(axis='y', alpha=0.3)

# 3. Points per line
axes[0, 2].hist(data['points_per_line'], bins=np.arange(0.5, max(data['points_per_line'])+1.5, 1), 
                edgecolor='black', alpha=0.7, color='mediumseagreen')
axes[0, 2].set_xlabel('Number of Points per Line')
axes[0, 2].set_ylabel('Frequency')
axes[0, 2].set_title(f'Points per Line\n(mean: {np.mean(data["points_per_line"]):.2f})')
axes[0, 2].grid(axis='y', alpha=0.3)

# Filter out zero-length error bars for visualization
nonzero_up = [x for x in data['error_bar_up'] if x > 0]
nonzero_down = [x for x in data['error_bar_down'] if x > 0]
nonzero_total = [x for x in data['error_bar_total'] if x > 0]

# 4. Error bar upper distances
axes[1, 0].hist(nonzero_up, bins=50, 
                edgecolor='black', alpha=0.7, color='purple')
axes[1, 0].set_xlabel('Upper Error Bar Distance (pixels)')
axes[1, 0].set_ylabel('Frequency')
axes[1, 0].set_title(f'Upper Error Bar Sizes (non-zero)\n(median: {np.median(nonzero_up):.1f}px, n={len(nonzero_up)})')
axes[1, 0].grid(axis='y', alpha=0.3)

# 5. Error bar lower distances
axes[1, 1].hist(nonzero_down, bins=50, 
                edgecolor='black', alpha=0.7, color='orange')
axes[1, 1].set_xlabel('Lower Error Bar Distance (pixels)')
axes[1, 1].set_ylabel('Frequency')
axes[1, 1].set_title(f'Lower Error Bar Sizes (non-zero)\n(median: {np.median(nonzero_down):.1f}px, n={len(nonzero_down)})')
axes[1, 1].grid(axis='y', alpha=0.3)

# 6. Total error bar length
axes[1, 2].hist(nonzero_total, bins=50, 
                edgecolor='black', alpha=0.7, color='crimson')
axes[1, 2].set_xlabel('Total Error Bar Length (pixels)')
axes[1, 2].set_ylabel('Frequency')
axes[1, 2].set_title(f'Total Error Bar Lengths (non-zero)\n(median: {np.median(nonzero_total):.1f}px)')
axes[1, 2].grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.savefig('real_distributions.png', dpi=150, bbox_inches='tight')
plt.show()

print("\nDistribution Summary:")
print(f"  Lines per image:    {np.min(data['lines_per_image'])} - {np.max(data['lines_per_image'])}")
print(f"  Points per image:   {np.min(data['points_per_image'])} - {np.max(data['points_per_image'])}")
print(f"  Points per line:    {np.min(data['points_per_line'])} - {np.max(data['points_per_line'])}")
if nonzero_up:
    print(f"  Upper bar size:     {np.min(nonzero_up):.1f} - {np.max(nonzero_up):.1f} px")
    print(f"  Lower bar size:     {np.min(nonzero_down):.1f} - {np.max(nonzero_down):.1f} px")
```

## 5. Image Quality Analysis

```python
# Analyze image quality: background noise, grid presence, contrast
brightness = []
contrast = []
edge_density = []

sample_size = min(50, len(image_files))
sample_files = random.sample(image_files, sample_size)

for img_file in tqdm(sample_files, desc='Analyzing quality'):
    img = cv2.imread(os.path.join(IMAGES_DIR, img_file))
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    # Brightness (mean intensity)
    brightness.append(gray.mean())
    
    # Contrast (std of intensity)
    contrast.append(gray.std())
    
    # Edge density (Canny edge count)
    edges = cv2.Canny(gray, 50, 150)
    edge_density.append(edges.sum() / edges.size)

fig, axes = plt.subplots(1, 3, figsize=(16, 4))

axes[0].hist(brightness, bins=20, edgecolor='black', alpha=0.7, color='gold')
axes[0].set_xlabel('Mean Brightness (0-255)')
axes[0].set_ylabel('Frequency')
axes[0].set_title(f'Brightness Distribution\n(mean: {np.mean(brightness):.1f})')
axes[0].grid(axis='y', alpha=0.3)

axes[1].hist(contrast, bins=20, edgecolor='black', alpha=0.7, color='teal')
axes[1].set_xlabel('Contrast (std of intensity)')
axes[1].set_ylabel('Frequency')
axes[1].set_title(f'Contrast Distribution\n(mean: {np.mean(contrast):.1f})')
axes[1].grid(axis='y', alpha=0.3)

axes[2].hist(edge_density, bins=20, edgecolor='black', alpha=0.7, color='indianred')
axes[2].set_xlabel('Edge Density (normalized)')
axes[2].set_ylabel('Frequency')
axes[2].set_title(f'Edge Density\n(mean: {np.mean(edge_density):.3f})')
axes[2].grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.savefig('real_quality.png', dpi=150, bbox_inches='tight')
plt.show()

print(f"\nImage Quality:")
print(f"  Brightness: {np.mean(brightness):.1f} ± {np.std(brightness):.1f} (higher = whiter background)")
print(f"  Contrast: {np.mean(contrast):.1f} ± {np.std(contrast):.1f} (higher = more variation)")
print(f"  Edge density: {np.mean(edge_density):.3f} ± {np.std(edge_density):.3f} (higher = more complex)")
```

## 6. Visual Examples (Various Styles)

```python
# Display diverse examples
sample_ids = random.sample(data['image_ids'], min(9, len(data['image_ids'])))

fig, axes = plt.subplots(3, 3, figsize=(18, 18))
axes = axes.flatten()

for idx, img_id in enumerate(sample_ids):
    # Load image
    img = Image.open(os.path.join(IMAGES_DIR, f"{img_id}.png"))
    
    # Load label
    with open(os.path.join(LABELS_DIR, f"{img_id}.json"), 'r') as f:
        label = json.load(f)
    
    # Display
    axes[idx].imshow(img)
    
    # Overlay
    for line in label:
        for point in line['points']:
            if point.get('label', '') == '':  # actual data point
                x, y = point['x'], point['y']
                dy_up = point.get('topBarPixelDistance', 0)
                dy_down = point.get('bottomBarPixelDistance', 0)
                
                # Draw
                axes[idx].plot(x, y, 'ro', markersize=6, markeredgecolor='white', markeredgewidth=1)
                
                if dy_up > 0 or dy_down > 0:
                    axes[idx].plot([x, x], [y - dy_up, y + dy_down], 'b-', linewidth=2, alpha=0.7)
                    if dy_up > 0:
                        axes[idx].plot(x, y - dy_up, 'b_', markersize=8)
                    if dy_down > 0:
                        axes[idx].plot(x, y + dy_down, 'b_', markersize=8)
    
    num_lines = len(label)
    num_points = sum(len([p for p in line['points'] if p.get('label', '') == '']) for line in label)
    axes[idx].set_title(f"{num_lines} line(s), {num_points} point(s)", fontsize=10)
    axes[idx].axis('off')

plt.tight_layout()
plt.savefig('real_examples.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 7. Domain Gap: Synthetic vs Real Comparison

```python
# Load synthetic data for comparison
synth_labels_dir = "/kaggle/input/med-synthetic-dataset/kaggle/working/dataset/labels"
synth_label_files = [f for f in os.listdir(synth_labels_dir) if f.endswith('.json')]

# Sample synthetic data
synth_sample = random.sample(synth_label_files, min(500, len(synth_label_files)))
synth_up = []
synth_down = []

for label_file in synth_sample:
    with open(os.path.join(synth_labels_dir, label_file), 'r') as f:
        label = json.load(f)
    for line in label:
        for point in line['points']:
            if point.get('label', '') == '':
                synth_up.append(point.get('topBarPixelDistance', 0))
                synth_down.append(point.get('bottomBarPixelDistance', 0))

# Compare distributions
fig, axes = plt.subplots(1, 2, figsize=(16, 5))

# Upper bars
axes[0].hist(synth_up, bins=50, alpha=0.6, label='Synthetic', color='blue', edgecolor='black')
axes[0].hist(nonzero_up, bins=50, alpha=0.6, label='Real', color='red', edgecolor='black')
axes[0].set_xlabel('Upper Error Bar Distance (px)')
axes[0].set_ylabel('Frequency (normalized)')
axes[0].set_title('Upper Error Bar Size Distribution')
axes[0].legend()
axes[0].grid(axis='y', alpha=0.3)

# Lower bars
axes[1].hist(synth_down, bins=50, alpha=0.6, label='Synthetic', color='blue', edgecolor='black')
axes[1].hist(nonzero_down, bins=50, alpha=0.6, label='Real', color='red', edgecolor='black')
axes[1].set_xlabel('Lower Error Bar Distance (px)')
axes[1].set_ylabel('Frequency (normalized)')
axes[1].set_title('Lower Error Bar Size Distribution')
axes[1].legend()
axes[1].grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.savefig('domain_gap_comparison.png', dpi=150, bbox_inches='tight')
plt.show()

print("\n" + "="*70)
print("DOMAIN GAP ANALYSIS")
print("="*70)
print(f"\nError Bar Sizes:")
print(f"  Synthetic - Upper: {np.median(synth_up):.1f}px (mean: {np.mean(synth_up):.1f})")
print(f"  Real - Upper:      {np.median(nonzero_up):.1f}px (mean: {np.mean(nonzero_up):.1f})")
print(f"  Synthetic - Lower: {np.median(synth_down):.1f}px (mean: {np.mean(synth_down):.1f})")
print(f"  Real - Lower:      {np.median(nonzero_down):.1f}px (mean: {np.mean(nonzero_down):.1f})")

print(f"\nKey Differences:")
print(f"  ⚠️ Real data has {data['zero_bars']} points with zero-length bars")
print(f"  ⚠️ Real images have {len(set(resolutions))} unique resolutions (synthetic: more uniform)")
print(f"  ⚠️ Real plots show more variation in styles, grids, backgrounds")
print(f"  ⚠️ Real error bars: wider range ({np.min(nonzero_up):.0f}-{np.max(nonzero_up):.0f}px vs synthetic: ~10-80px)")
```
